# **3. Data Cleaning**

```python
df.isnull().sum()
```

```python
# Remove all of unnecessary attributes
df.drop(columns=['Accident_Index','Location_Easting_OSGR','Location_Northing_OSGR','Local_Authority_(District)','Local_Authority_(Highway)','1st_Road_Class','1st_Road_Number','Junction_Control','2nd_Road_Class','2nd_Road_Number','Special_Conditions_at_Site','Carriageway_Hazards','LSOA_of_Accident_Location'],inplace=True)
```

```python
df.info()
```

```python
# Replace all the Null Values with Median Values
df['Pedestrian_Crossing-Human_Control'].fillna(df['Pedestrian_Crossing-Human_Control'].mode()[0],inplace=True)
df['Pedestrian_Crossing-Physical_Facilities'].fillna(df['Pedestrian_Crossing-Physical_Facilities'].mode()[0],inplace=True)
df['Longitude'].fillna(df['Longitude'].mode()[0],inplace=True)
df['Time'].fillna(df['Time'].mode()[0],inplace=True)
df.isnull().sum()
```

# **Exploratery Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **1.1 Number of Accident on different types of Road**

```python
plt.figure(figsize=(10, 5))
road = df['Road_Type'].value_counts()
road.plot(kind='bar', color='r',edgecolor='black')

plt.title(('Accident on different types of Road'),fontsize=20,fontweight='bold')
plt.xlabel(('Road Type'),fontsize=15,fontweight='bold')
plt.ylabel(('Number of Accidents'),fontsize=15,fontweight='bold')

plt.xticks(rotation=0, fontsize=12, ha='center') #'ha' is used for writing the text in Left, Right, or Center
plt.yticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--')

# Add labels above each bar
for index, value in enumerate(road):
    plt.text(index, value + 0.5, str(value), ha='center', fontsize=11)
plt.tight_layout()  # Adjusts the plot to ensure everything fits without overlap
plt.show()
```

# **1.2 Accident due to different Surfaces of Road**

```python
plt.figure(figsize=(10, 5))
road=df['Road_Surface_Conditions'].value_counts()
road.plot(kind='bar',color='r',edgecolor='black')

plt.title(('Accident due to Road Surface'),fontsize=20,fontweight='bold')
plt.xlabel(('Road Surface'),fontsize=15,fontweight='bold')
plt.ylabel(('Number of Accidents'),fontsize=15,fontweight='bold')

plt.xticks(rotation=0,fontsize=12,ha='center')
plt.yticks(rotation=0,fontsize=12)
plt.grid(axis='y',linestyle='--')

# Add labels above each bar, below 2 lines is used for calculating the total numbers
for index, value in enumerate(road):
    plt.text(index, value + 0.5, str(value), ha='center', fontsize=11)
plt.tight_layout()
plt.show()
```

# **1.3 Accident Severity by Road Surface Conditions**

```python
plt.figure(figsize=(12, 6))
road_surface_severity = df.groupby(['Road_Surface_Conditions', 'Accident_Severity']).size().unstack()
road_surface_severity.plot(kind='bar', stacked=True) #if we change "stacked=False" then baar graph not overlap on each other

plt.title(('Accident Severity by Road Surface Conditions'),fontsize=17,fontweight='bold')
plt.xlabel(('Road Surface Conditions'),fontsize=15,fontweight='bold')
plt.ylabel(('Number of Accidents'),fontsize=15,fontweight='bold')

plt.xticks(rotation=45, fontsize=10, ha='right')
plt.yticks(rotation=0, fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.legend(title='Accident Severity')
plt.show()
```

# **4.3 Number of Accident by Hour of Day**

```python
plt.figure(figsize=(10, 5))
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour # Use '%H:%M' to match the time format in your data
sns.histplot(df['Hour'],bins=24, kde=False,  color='c')

plt.title(('Accidents by Hour of Day'),fontsize=18,fontweight='bold')
plt.xlabel(('Hour of Day'),fontsize=15,fontweight='bold')
plt.ylabel(('Number of Accidents'),fontsize=15,fontweight='bold')

plt.xticks(range(1, 25),ha='center')
plt.yticks(rotation=0,fontsize=12)
plt.grid(axis='y',linestyle='--')

plt.tight_layout()
plt.show()
```

# **5.1 Urban vs. Rural Analysis**
